### Cleaning Race and Sprint Race Data

Below we will clean up the Race and Sprint Race Data, while also merging circuit info through the matching circuit ID, adding location data such as city, country, latitude, and longitude. This will be for the purposes of visualizing on a map just extensive the schedule is in terms of traveling.

```python
races.tail()
```

```python
race_data = races.drop(columns=['url', 'time', 'fp1_date', 'fp1_time', "fp2_date", 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date', 'quali_time', 'sprint_date', 'sprint_time'])
```

```python
race_data.head()
```

```python
circuit_info.head()
```

```python
race_data = pd.merge(race_data, circuit_info[['circuitId','location', 'country', 'lat', 'lng']], on='circuitId', how='left')
```

```python
race_data.head(5)
```

```python
sprint_data = races[['raceId', 'year', 'round', 'circuitId', 'name', 'date', 'sprint_date']]
```

```python
sprint_data.head()
```

```python
sprint_data = sprint_data[sprint_data['sprint_date'] != '\\N']
```

```python
sprint_data = pd.merge(sprint_data, circuit_info[['circuitId','location', 'country', 'lat', 'lng']], on='circuitId', how='left')
```
