2023 Race IDs = 1098-1120

### Narrow down driver standings by raceId to 2023 season.

```python
driver_standings_2023 = driver_standings[(driver_standings["raceId"] >= 1098) & (driver_standings["raceId"]<= 1120)]
```

```python
driver_standings_2023
```

```python
# Merge the driver info with the 2023 standings
driver_standings_2023_names = driver_standings_2023.merge(driver_info[['driverId','code','forename', 'surname']], on='driverId')
```

```python
driver_standings_2023_names
```

```python
# Merge the previous DF with name of the race
driver_standings_2023_races = driver_standings_2023_names.merge(races_2023[['raceId', 'name']], on='raceId')
```

```python
driver_standings_2023_races
```

```python
driver_standings_2023_races = driver_standings_2023_races.rename(columns={'name':'circuitName'})
```

```python
driver_standings_2023_races
```

```python
sorted_races = driver_standings_2023_races.sort_values(by=['raceId', 'position'])
```
